1. Explain the key features of Python that make it a popular choice for programming.

Answer: Python is widely popular due to its:

Simplicity: Easy-to-read syntax similar to English.

Versatility: Supports multiple paradigms (object-oriented, procedural, functional).

Extensive Libraries: Includes libraries for data analysis, machine learning, web development, etc.

Portability: Runs on multiple platforms (Windows, Linux, macOS).

Dynamic Typing: Variables are not bound to a specific type.

3. Compare and contrast mutable and immutable objects in Python with examples.

Answer:
Mutable Objects: Can be modified after creation (e.g., lists, dictionaries).
Immutable Objects: Cannot be changed after creation (e.g., tuples, strings).

Example:
# Mutable
lst = [1, 2, 3]
lst[0] = 10
print(lst)  # Output: [10, 2, 3]

# Immutable
tup = (1, 2, 3)
# tup[0] = 10  # Error: 'tuple' object does not support item assignment

4. Discuss the different types of operators in Python and provide examples of how they are used.

Answer:
Python operators include:

Arithmetic Operators: +, -, *, /

print(3 + 2)  # Output: 5
Relational Operators: >, <, ==

print(5 > 3)  # Output: True
Logical Operators: and, or, not

print(True and False)  # Output: False
Assignment Operators: =, +=, -=

x = 5
x += 3

Membership Operators: in, not in

print('a' in 'apple')  # Output: True

5. Explain the concept of type casting in Python with examples.

Answer:
Type casting is the process of converting a variable from one type to another.

Example:


# Implicit casting
x = 5
y = 2.0
z = x + y  # Python converts 'x' to float
print(z)  # Output: 7.0

# Explicit casting
num = "10"
print(int(num) + 5)  # Output: 15
6. How do conditional statements work in Python? Illustrate with examples.

Answer:
Conditional statements control the flow of execution based on conditions using if, elif, and else.

Example:

x = 10
if x > 5:
    print("x is greater than 5")
elif x == 5:
    print("x is 5")
else:
    print("x is less than 5")
Output:
x is greater than 5
7. Describe the different types of loops in Python and their use cases with examples.

Answer:

for Loop: Iterates over a sequence.
for i in range(3):
    print(i)  # Output: 0, 1, 2
while Loop: Runs until a condition is false.
count = 0
while count < 3:
    print(count)
    count += 1  # Output: 0, 1, 2
Use Cases:
Use for for fixed iterations.
Use while for indefinite loops based on a condition.
