1. What are data structures, and why are they important?
Ans. Data structures are organized ways to store, manage, and access data efficiently. They are important because they:

Enable efficient data retrieval and manipulation.

Support algorithmic solutions to computational problems.

Optimize storage and processing time in applications such as databases, operating systems, and AI models.

2. Explain the difference between mutable and immutable data types with examples.

Ans. Mutable: These can be changed after creation. For example, a Python list:

lst = [1, 2, 3]
lst[0] = 10  # Modifies the list
print(lst)  # Output: [10, 2, 3]

Immutable: These cannot be changed once created. For example, a Python tuple:

tpl = (1, 2, 3)
tpl[0] = 10  # Raises an error

3. What are the main differences between lists and tuples in Python?

Ans. Lists: Mutable, slower, and used for dynamic collections. Defined with square brackets [].

Tuples: Immutable, faster, and used for fixed collections. Defined with parentheses ().
Example:

lst = [1, 2, 3]  # List
tpl = (1, 2, 3)  # Tuple

4. Describe how dictionaries store data.
Ans. Dictionaries store data as key-value pairs. Internally, they use a hash table where keys are hashed to determine their storage location, ensuring fast retrieval based on the key.
Example:

my_dict = {"name": "Alice", "age": 25}
print(my_dict["name"])  # Output: Alice

5. Why might you use a set instead of a list in Python?

Ans. To ensure uniqueness: Sets automatically eliminate duplicate elements.

For faster membership tests: Checking for an element in a set is faster (average time complexity O(1)) compared to a list (O(n)).

6. What is a string in Python, and how is it different from a list?

Ans. A string is a sequence of characters and is immutable.

A list is mutable and can store elements of any data type.
Example:

str_val = "hello"
list_val = ["h", "e", "l", "l", "o"]

7. How do tuples ensure data integrity in Python?
Ans. Tuples are immutable, meaning their content cannot be changed after creation. This immutability ensures that data remains consistent and prevents accidental modifications.

8. What is a hash table, and how does it relate to dictionaries in Python?
Ans. A hash table is a data structure that maps keys to values using a hash function. Python dictionaries use hash tables for fast key-based data retrieval.

9. Can lists contain different data types in Python?
Ans. Yes, Python lists can hold elements of various data types, such as integers, strings, and objects.
Example:

lst = [1, "hello", 3.14]

10. Explain why strings are immutable in Python.
Ans. Strings are immutable to:

Ensure thread safety during concurrent operations.

Allow safe sharing of string objects in memory.

11. What advantages do dictionaries offer over lists for certain tasks?
Ans. Dictionaries:

Provide fast lookups by key (O(1) average time complexity).

Allow meaningful associations through key-value pairs, unlike lists which are index-based.

12. Describe a scenario where using a tuple would be preferable over a list.
Ans. Use tuples when you need fixed, read-only collections, such as coordinates in a 3D space:

point = (10, 20, 30)  # Represents (x, y, z)

13. How do sets handle duplicate values in Python?
Ans. Sets automatically remove duplicate values, ensuring all elements are unique.
Example:

s = {1, 2, 2, 3}
print(s)  # Output: {1, 2, 3}

14. How does the “in” keyword work differently for lists and dictionaries?

Ans. In lists: Checks if a value exists in the list.

In dictionaries: Checks if a key exists.
Example:

lst = [1, 2, 3]
print(2 in lst)  # True

my_dict = {"key": "value"}
print("key" in my_dict)  # True

15. Can you modify the elements of a tuple? Explain why or why not.
Ans. No, tuples are immutable, so their elements cannot be changed after creation. This characteristic ensures data integrity.

16. What is a nested dictionary, and give an example of its use case.
Ans. A nested dictionary is a dictionary inside another dictionary. It’s used for storing hierarchical or grouped data.
Example:

student = {
    "name": "Alice",
    "scores": {"math": 90, "science": 85}
}

17. Describe the time complexity of accessing elements in a dictionary.
Ans. Accessing elements in a dictionary has an average time complexity of O(1) due to its hash table implementation.

18. In what situations are lists preferred over dictionaries?
Ans. Lists are better when:

The data needs to maintain order.

Sequential operations are required.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
Ans. Before Python 3.7, dictionaries didn’t maintain insertion order, which made data retrieval order unpredictable. From Python 3.7 onward, dictionaries preserve insertion order.

20. Explain the difference between a list and a dictionary in terms of data retrieval.

Ans. Lists retrieve elements by index (sequential storage).

Dictionaries retrieve values by keys (key-value mapping).
